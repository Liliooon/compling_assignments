### Задание 1

```python
import re
```

```python
with open('zhivago.txt', encoding='utf-8') as f:
    text = f.read()
```

```python
def remove_trash(text):
    no_tags_text = re.sub(r'<[^>]+>', ' ', text)
    no_space_sequences_text = re.sub(r'[\s\\xa0]+', ' ', no_tags_text).strip()
    text_till_last_sentence = f"{no_space_sequences_text.rsplit('.', 1)[0]}."
    return text_till_last_sentence
```

```python
text = remove_trash(text)
```

```python
text[:1500]
```

### Задание 2

```python
from collections import Counter
from string import punctuation
from rusenttokenize import ru_sent_tokenize
from razdel import tokenize as razdel_tokenize
```

```python
sentences = ru_sent_tokenize(text)
```

```python
sentences[200:210]
```

```python
text = text.lower()
```

```python
punctuation = f'{punctuation}«»–...'
```

```python
tokens = [token.text.strip(punctuation) for token in razdel_tokenize(text) 
          if token.text not in punctuation]
tokens = [token for token in tokens if token]
```

```python
tokens[:10]
```

```python
# есть ли в тексте повторяющиеся корректные предложения? если да то какие?
```

```python
def sorted_by_freq(lst):
    counted_lst = dict(Counter(lst)).items()
    sorted_by_freq_counted_lst = sorted(counted_lst, key=lambda x: x[1], reverse=True)
    return sorted_by_freq_counted_lst
```

```python
[sent for sent in sorted_by_freq(sentences) if sent[1] > 1]
```

```python
# какой самый частотный токен в тексте длиннее 6 символов?
```

```python
[word for word in sorted_by_freq(tokens) if len(word[0]) > 6][0]
```

### Задание 3

```python
from nltk.stem.snowball import SnowballStemmer
```

```python
stemmer = SnowballStemmer('russian')
```

```python
stems = [(token, stemmer.stem(token)) for token in tokens]
```

```python
# два разных слова ошибочно свелись к одинаковой основе
```

```python
same_stems_for_diff_tokens = [(stem, another_stem) for stem in stems for another_stem in list(set(stems))
                              if stem[0] != another_stem[0] and stem[1] == another_stem[1]]
```

```python
[pair for pair in sorted_by_freq(same_stems_for_diff_tokens)][:40]
```

В первых трех из них у второй пары внутри каждой пары основа должна быть нулевой, а в четвертой и пятой парах у вторых пар основа должна быть одинакова с токеном.

```python
# слово не изменилось после стемминга (слово должно быть русским и длиннее 4 символов)
```

```python
unchanged_tokens = [stem for stem in stems if stem[0] == stem[1] and 
                    len(stem[0]) > 4 and re.search(r'[а-яё]', stem[0])]
```

```python
[token for token in sorted_by_freq(unchanged_tokens)][:15]
```

### Задание 5

```python
from stop_words import get_stop_words
from pymystem3 import Mystem
import pymorphy2
```

```python
stop_words = set(get_stop_words('ru'))
```

```python
# лемматизируйте токены с помощью pymorphy2
```

```python
morph = pymorphy2.MorphAnalyzer()
```

```python
def get_pymorphy2_lemmatized_text(tokens):
    lemmatized_text = [morph.parse(token)[0].normal_form for token in tokens]
    lemmatized_text_without_stops = [lemma for lemma in lemmatized_text 
                                     if re.search('[а-яёa-z]', lemma) and lemma not in stop_words]
    return lemmatized_text_without_stops
```

```python
pymorphy2_lemmatized_text = get_pymorphy2_lemmatized_text(tokens)
```

```python
# лемматизируйте текст с помощью mystem3 
```

```python
m = Mystem()
```

```python
def get_mystem_lemmatized_text(text):
    lemmatized_text = [lemma for lemma in m.lemmatize(text)]
    lemmatized_text_without_stops = [lemma for lemma in lemmatized_text 
                                     if re.search('[а-яёa-z]', lemma) and lemma not in stop_words]
    return lemmatized_text_without_stops
```

```python
mystem_lemmatized_text = get_mystem_lemmatized_text(' '.join(tokens))
```

```python
sorted_by_freq(pymorphy2_lemmatized_text)
```

```python
sorted_by_freq(mystem_lemmatized_text)
```
